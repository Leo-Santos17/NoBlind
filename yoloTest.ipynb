```python
from ultralytics import YOLO
```

# Modelo

```python
model = YOLO("yolo11n.pt")
```

# Imagem de uso

```python
img = "palestra.png"
```

# Execução

```python
results = model(img)
```

```python
for result in results:
    boxer = result.boxes
    result.save(filename="result.jpg")
```

# Mostrar resultado

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
img = 'result.jpg'
image = mpimg.imread(img)
plt.imshow(image)
plt.axis('off')
plt.show()
```
